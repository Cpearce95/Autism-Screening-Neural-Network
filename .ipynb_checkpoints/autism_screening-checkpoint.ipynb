```python
import sys
import pandas as pd
import sklearn
import keras

print 'Python: {}'.format(sys.version)
print 'Pandas: {}'.format(pd.__version__)
print 'Sklearn: {}'.format(sklearn.__version__)
print 'Keras: {}'.format(keras.__version__)
```

```python
#Importing the autism data
file = 'Autism-Child-Data.txt'

#Reading the file into the notebook
data = pd.read_csv(file, sep=',', index_col=None)
```

```python
print 'Shape of DataFrame: {}'.format(data.shape)
print data.loc[0]
```

```python
#Dropping unwanted columns
data = data.drop(['result', 'age_desc'], axis=1)
# print out first 11 patients
data.loc[:10]
```

```python
# Printing out a dataframe description
data.describe()
```

```python
data.dtypes
```

```python
# Creating X (used for prediction) Y (Whether they have Autism) sets for training.
prediction = data.drop(['class/ASD'], 1)
autismDiagnosis = data['class/ASD']
```

```python
# Printing out x table
prediction.loc[:10]
```

```python
predictionWithDummies = pd.get_dummies(prediction)
```

```python
predictionWithDummies.loc[:10]
```

```python
predictionWithDummies.columns.values
```

```python
# Printing out an example for one patient
predictionWithDummies.loc[0]
```

```python
# Converting the class data to categorical values
autismDiagnosisDummies = pd.get_dummies(autismDiagnosis)
```

```python
autismDiagnosisDummies.iloc[:10]
```

```python
from sklearn import model_selection
# splite the X, Y data into training and testing datasets
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(predictionWithDummies, autismDiagnosisDummies, test_size = 0.2)
```

```python
print X_train.shape
print X_test.shape
print Y_train.shape
print Y_test.shape
#END OF DATA PRE-PROCESSING PHASE
```

```python
# Building a neural network using Keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam

#Defining a function to build the Keras model
def create_model():
    #create model
    model = Sequential()
    model.add(Dense(8, input_dim = 96, kernel_initializer = 'normal', activation = 'relu'))
    model.add(Dense(4, kernel_initializer = 'normal', activation = 'relu'))
    model.add(Dense(2, activation = 'sigmoid'))
    
    #compile the model
    adam = Adam(lr = 0.001)
    model.compile(loss = 'categorical_crossentropy', optimizer = adam, metrics = ['accuracy'])
    
    return model
```

```python
#Create the model
model = create_model()
print(model.summary())
```

```python
# Fitting the model to the training data
model.fit(X_train, Y_train, epochs = 50, batch_size = 10, verbose = 1)
```

```python
# Generating a classification report using predictions from the model
from sklearn.metrics import classification_report, accuracy_score

predictions = model.predict_classes(X_test)
```

```python
predictions
```

```python
print('Prediction Results for Neural Network')
print(accuracy_score(Y_test[['YES']], predictions))
print(classification_report(Y_test[['YES']], predictions))
```
